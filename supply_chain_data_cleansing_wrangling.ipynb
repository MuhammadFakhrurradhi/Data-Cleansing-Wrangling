```python
# after checking the data structure, some missing values and unuseful data columns are found.
# those columns are dropped.
data = data.drop(["Customer Email","Customer Password", "Product Image","Order Zipcode","Product Description", "Product Status","Customer Zipcode"], axis = 1)

data.head(5)
```

```python
# creating function for renaming columns name

def rename_column(data, initial, final):
    """
    Rename a column in DataFrame.
    
    Args:
        data: DataFrame containing the column to be rename.
        initial (str) : The current column name.
        final (str) : The new column name.
        
    Returns : 
        data: DataFrmae with the column renamed.
    """
    data.rename(columns={initial: final}, inplace = True)
    return data.head(5)
```

```python
# define function to format columns into title formatting
def title_string(data, string):
    data[string] = data[string].str.title()
    return data.head(5)
```

```python
#calling the function

title_string(data,"transaction_type")
title_string(data,"delivery_status")
title_string(data, "order_status")
```

```python
# define function to replace unwanted character/information within columns.
def replace(data, string, initial, replacing):
    data[string] = data[string].str.replace(initial,replacing)
    return data.head(5)
```

```python
# calling the function
replace(data, "order_status", "_"," ")

# "EE. UU." known as United States in spanish.
replace(data, "customer_country", "EE. UU.","United States")
```

```python
# Some columns that representing currency are having inconsitency decimal. 
#defining function to round up the decimal.
def round_decimal(data, column):
    data[column] = data[column].apply(lambda x: round(x,2))
    return data.head(5)
```

```python
# calling the function
round_decimal(data, "profit_per_order")
round_decimal(data, "sales_per_customer")
round_decimal(data, "order_item_discount")
round_decimal(data, "order_item_total")
round_decimal(data, "order_profit_per_order")
round_decimal(data, "order_item_discount_rate")
round_decimal(data, "order_item_product_price")
round_decimal(data, "order_item_profit_ratio")
round_decimal(data, "sales")
round_decimal(data, "product_price")
```
